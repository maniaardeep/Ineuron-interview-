### Q: What is the purpose of a technical skills review?

**Answer:**
A technical skills review evaluates a candidate’s practical knowledge in programming, problem-solving, and system understanding. 
It helps employers assess whether the person can apply concepts in real-world scenarios, write clean code, and think logically.

### Q: Explain the difference between software testing and software debugging.

**Answer:**
Software testing is the process of identifying bugs or errors in a system.
Software debugging is the process of fixing those identified bugs.
In short: Testing finds problems, debugging solves them.

### Q: Explain the concept of Big-O notation and its significance in algorithm analysis.

**Answer:**
Big-O notation measures the time or space complexity of an algorithm.
It tells how the performance of an algorithm grows as input size increases.

Example:
O(1) → constant time
O(n) → linear time
O(n^2) → quadratic time

It helps in comparing efficiency of algorithms.

### Q: What is the difference between procedural programming and object-oriented programming?

**Answer:**
Procedural Programming:
- Based on functions
- Follows step-by-step instructions

Object-Oriented Programming (OOP):
- Based on objects and classes
- Supports concepts like inheritance, encapsulation, polymorphism

OOP is more scalable and reusable.

### Q: Explain the concept of recursion in programming and provide an example.

**Answer:**
Recursion is a technique where a function calls itself to solve a problem.

Example (Factorial):
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

It breaks a problem into smaller subproblems.
