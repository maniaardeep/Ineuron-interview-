### Q: What is the purpose of ML interviews?

**Answer:**
ML interviews check whether a candidate can apply machine learning concepts to real-world problems.
They test understanding of algorithms, data handling, coding skills, and business thinking.
Companies want to see if you can solve problems using data, not just explain theory.

### Q: Explain the difference between supervised and unsupervised learning.

**Answer:**
Supervised learning uses labeled data where input-output pairs are known (e.g., classification, regression).
Unsupervised learning works with unlabeled data and tries to find patterns (e.g., clustering).
Example: Predicting house price = supervised, grouping customers = unsupervised.

### Q: What is overfitting in machine learning?

**Answer:**
Overfitting happens when a model learns training data too well, including noise, and performs poorly on new data.
It results in high training accuracy but low test accuracy.

### Q: What are the key components of a machine learning model?

**Answer:**
1. Data (input features)
2. Algorithm (model)
3. Loss function (error measurement)
4. Optimization method (learning process)
5. Evaluation metrics (performance measurement)

### Q: How do you handle missing data in a dataset?

**Answer:**
Common methods:
- Remove rows/columns
- Mean/median/mode imputation
- Model-based imputation
- Forward/backward fill (time series)

### Q: Explain the bias-variance tradeoff.

**Answer:**
Bias is error due to overly simple models.
Variance is error due to overly complex models.
Goal is to balance both to achieve good generalization.

### Q: What is cross-validation and why is it important?

**Answer:**
Cross-validation splits data into multiple folds and trains/tests multiple times.
It provides a more reliable estimate of model performance and reduces overfitting risk.

### Q: What is the curse of dimensionality?

**Answer:**
As the number of features increases, data becomes sparse and model performance degrades.
It makes learning harder and increases computational cost.

### Q: How do you evaluate a machine learning model's performance?

**Answer:**
Depends on problem:
- Classification: Accuracy, Precision, Recall, F1-score, ROC-AUC
- Regression: MAE, MSE, RMSE, R²

### Q: Difference between precision and recall?

**Answer:**
Precision = Correct positive predictions / Total predicted positives
Recall = Correct positive predictions / Actual positives
Precision focuses on false positives, Recall focuses on false negatives.

### Q: What is regularization and why is it important?

**Answer:**
Regularization adds penalty to model complexity.
It prevents overfitting.
Types:
- L1 (Lasso)
- L2 (Ridge)

### Q: What is feature engineering?

**Answer:**
Feature engineering is the process of creating, transforming, and selecting useful features from raw data.
It improves model performance significantly.

### Q: What are hyperparameters?

**Answer:**
Hyperparameters are parameters set before training (not learned).
Examples:
- Learning rate
- Number of trees
- Depth of tree

### Q: Difference between classification and regression?

**Answer:**
Classification predicts categories (e.g., spam or not).
Regression predicts continuous values (e.g., house price).
